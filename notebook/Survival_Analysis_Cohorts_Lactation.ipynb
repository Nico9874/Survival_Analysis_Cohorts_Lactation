# 2. Estimación de la Función de Supervivencia

## 2.1. Curva Kaplan-Meier Global

```r
# Crear el objeto de supervivencia
surv_obj <- Surv(time = data$survival, event = data$cens)

# Ajustar el modelo Kaplan-Meier
km_fit <- survfit(surv_obj ~ 1, data = data)

# Visualizar la curva de supervivencia
# Pass the 'data' argument to ggsurvplot
ggsurvplot(km_fit,
           data = data, # This line is added to pass the data explicitly
           conf.int = TRUE,         # Intervalos de confianza
           risk.table = TRUE,       # Tabla de riesgo debajo de la curva
           title = "Curva de Supervivencia Kaplan-Meier - Lactancia Materna",
           xlab = "Semanas",
           ylab = "Probabilidad de continuar lactando",
           ggtheme = theme_minimal())

# Probabilidades de continuar lactando después de 12, 24 y 36 semanas
summary(km_fit, times = c(12, 24, 36))
```

## 2.2. Comparación de Curvas por Raza

```r
# Ajustar el modelo Kaplan-Meier con Race como factor
km_fit_race <- survfit(Surv(survival, cens) ~ race, data = data)

# Visualizar las curvas de supervivencia por grupo racial
# Pass the data argument to ggsurvplot
ggsurvplot(km_fit_race, data = data,
           conf.int = TRUE,         # Mostrar intervalos de confianza
           pval = TRUE,             # Mostrar p-valor de la prueba log-rank
           risk.table = TRUE,       # Tabla de riesgo debajo del gráfico
           legend.title = "Raza",
           legend.labs = c("Blanca", "Negra", "Otra"),
           title = "Curvas de Supervivencia Kaplan-Meier por Raza",
           xlab = "Semanas",
           ylab = "Probabilidad de continuar lactando",
           ggtheme = theme_minimal())
```

# 3. Modelo de Riesgos Proporcionales de Cox

## 3.1. Ajuste del Modelo

```r
# Cargar la librería survival si no está instalada
if (!require(survival)) install.packages("survival", dependencies = TRUE)

# Cargar la librería
library(survival)

# Ajustar el modelo de Cox con las variables indicadas
cox_model <- coxph(Surv(survival, cens) ~ race + poverty + smoked + alcohol + age + education + prenatal, data = data)

# Mostrar el resumen del modelo
summary(cox_model)
```

## 3.2. Interpretación de Resultados

¿Qué factores aumentan o disminuyen el riesgo de finalizar la lactancia?
Al revisar los resultados del modelo, hay varios factores que parecen influir en cuánto tiempo una madre mantiene la lactancia. La raza es un factor importante: el resultado muestra que algunas madres tienen un 15% más de probabilidad de interrumpir la lactancia antes en comparación con el grupo de referencia. Esto es algo significativo, lo que quiere decir que no es un resultado al azar, sino que realmente hay una diferencia según la raza.

Otro factor clave es la pobreza, pero en este caso tiene un efecto opuesto: las madres en situación de pobreza tienen un 18% menos de probabilidad de interrumpir la lactancia. Es decir, parecen amamantar por más tiempo en comparación con aquellas que no están en situación de pobreza. Esto es interesante porque va en contra de lo que se podría suponer, y quizás se debe a razones económicas o culturales, como el acceso a la fórmula infantil o el valor que se le da a la lactancia en estos grupos.

Fumar y el consumo de alcohol también aparecen en el análisis, y aquí sí los resultados son más predecibles. Las madres que fuman tienen un 22% más de probabilidad de dejar de amamantar antes, lo cual tiene sentido porque el tabaco puede afectar la producción de leche y la decisión de continuar con la lactancia. En cuanto al alcohol, parece aumentar el riesgo en un 20%, pero este resultado no es tan sólido porque no es estadísticamente significativo, lo que significa que podría ser una coincidencia y no un efecto real.

Otro punto interesante es la educación. Según el modelo, las madres con más años de educación prolongan más la lactancia. De hecho, por cada año extra de educación, el riesgo de interrumpir la lactancia baja un 6%. Esto podría estar relacionado con el acceso a información sobre la importancia de la lactancia materna y el impacto positivo que tiene en la salud del bebé.

¿Qué variables no tienen un efecto claro?
No todas las variables en el modelo parecen afectar la duración de la lactancia. Por ejemplo, la edad de la madre no muestra un efecto significativo, aunque en el modelo el coeficiente es positivo. Esto significa que, aunque en teoría podría haber una relación entre la edad y la duración de la lactancia, en este análisis no hay suficiente evidencia para decir que las madres mayores amamantan por menos tiempo.

Otra variable que sorprendentemente no tiene un impacto fuerte es la atención prenatal. El resultado muestra que recibir atención prenatal tardía no cambia mucho el riesgo de interrumpir la lactancia, ya que el hazard ratio es casi 1 (es decir, no hay diferencia real entre quienes tuvieron o no atención prenatal temprana). Esto podría deberse a que la lactancia está influenciada por muchos otros factores que no dependen solo del control médico durante el embarazo.

Diferencias en la duración de la lactancia según la raza
Un hallazgo importante es que la raza sí parece influir en la duración de la lactancia. El modelo muestra que algunas madres tienen un 15% más de probabilidad de interrumpir la lactancia antes, y esto es significativo. ¿Por qué ocurre esto? Podría estar relacionado con diferencias culturales, apoyo a la lactancia en distintos grupos raciales o incluso diferencias en el acceso a información y asesoramiento sobre lactancia. Este es un punto clave que vale la pena explorar más a fondo.

En resumen, hay varios factores que influyen en la duración de la lactancia. Otros factores importantes que aparecen en el modelo de Cox incluyen:

* Pobreza: Las madres en condición de pobreza tienen un 17% menos de probabilidad de interrumpir la lactancia en comparación con las que no están en pobreza. Esto es interesante porque sugiere que las madres con menos recursos pueden estar más incentivadas a prolongar la lactancia por razones económicas o culturales.

* Fumar: Las madres que fuman tienen un 23% más de probabilidad de terminar la lactancia antes, lo cual es bastante preocupante porque el tabaco puede afectar tanto la producción de leche como la salud del bebé.

* Consumo de alcohol: Aunque el modelo sugiere que el alcohol podría aumentar el riesgo de interrumpir la lactancia en un 20%, el resultado no es significativo, por lo que no podemos asegurarlo con certeza.

* Educación: Las madres con más años de educación tienen un 6% menos de riesgo de interrumpir la lactancia por cada año adicional de estudios. Esto sugiere que el acceso a información y educación sobre los beneficios de la lactancia juega un papel clave en su duración.

* Edad y atención prenatal: Sorprendentemente, la edad de la madre y la atención prenatal no tienen un impacto significativo en la duración de la lactancia. Aunque podríamos pensar que el control prenatal influiría en la decisión de amamantar, los datos no muestran una relación clara.

## 4.2. Recomendaciones de Políticas de Salud

**Programas educativos sobre lactancia materna**

Dado que las madres con más educación tienden a amamantar por más tiempo, sería importante desarrollar programas informativos sobre lactancia dirigidos a todas las madres, especialmente aquellas con menos acceso a educación formal. Esto podría incluir charlas en centros de salud, hospitales y redes de apoyo comunitarias.

**Estrategias para reducir el consumo de tabaco en madres lactantes**

El modelo muestra que el tabaquismo está fuertemente asociado con una menor duración de la lactancia. Se podrían implementar campañas dirigidas a madres fumadoras, brindando apoyo para dejar el tabaco durante el embarazo y el postparto. También sería útil que los profesionales de salud informaran sobre cómo el tabaco afecta la producción de leche.

**Apoyo social y económico para madres en pobreza**

Las madres en pobreza parecen amamantar por más tiempo, posiblemente porque es una alternativa más accesible en términos económicos. Sin embargo, se debe garantizar que todas las madres, sin importar su nivel socioeconómico, tengan acceso a información y apoyo adecuado para evitar que la lactancia prolongada en estos casos sea solo por necesidad y no por elección informada.

**Mejorar el acceso a redes de apoyo para la lactancia**

Si bien la atención prenatal no parece influir directamente en la duración de la lactancia, el acceso a redes de apoyo sí podría ser clave. Se podrían fortalecer los grupos de lactancia en hospitales, centros de salud y comunidades, para que las madres reciban orientación y respaldo durante todo el proceso de lactancia.

**Conclusión Final**

En general, este análisis muestra que la lactancia materna se ve afectada por varios factores como la educación, el tabaquismo, la pobreza y la raza. Aunque la edad y la atención prenatal no parecen tener un impacto claro, sí hay espacio para mejorar el apoyo a las madres en diferentes áreas. Para prolongar la lactancia materna, se deberían impulsar estrategias educativas, apoyo a madres fumadoras y un mayor acceso a información en comunidades más vulnerables.

Estos hallazgos pueden servir como base para diseñar políticas públicas enfocadas en mejorar la salud materno-infantil y fomentar la lactancia materna de manera más efectiva.
